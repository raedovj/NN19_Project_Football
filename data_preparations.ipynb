# trying to form a table that could be used for some machine_learning algorithm

```python
# dropping all the players as they mostly seem to be null
final_df = match_t.drop(['home_player_X1', 'home_player_X2', 'home_player_X3', 'home_player_X4', 
                         'home_player_X5', 'home_player_X6', 'home_player_X7', 'home_player_X8', 
                         'home_player_X9', 'home_player_X10', 'home_player_X11', 'away_player_X1', 
                         'away_player_X2', 'away_player_X3', 'away_player_X4', 'away_player_X5', 
                         'away_player_X6', 'away_player_X7', 'away_player_X8', 'away_player_X9', 
                         'away_player_X10', 'away_player_X11', 'home_player_Y1', 'home_player_Y2', 
                         'home_player_Y3', 'home_player_Y4', 'home_player_Y5', 'home_player_Y6', 
                         'home_player_Y7', 'home_player_Y8', 'home_player_Y9', 'home_player_Y10', 
                         'home_player_Y11', 'away_player_Y1', 'away_player_Y2', 'away_player_Y3', 
                         'away_player_Y4', 'away_player_Y5', 'away_player_Y6', 'away_player_Y7', 
                         'away_player_Y8', 'away_player_Y9', 'away_player_Y10', 'away_player_Y11', 
                         'home_player_1', 'home_player_2', 'home_player_3', 'home_player_4', 
                         'home_player_5', 'home_player_6', 'home_player_7', 'home_player_8', 
                         'home_player_9', 'home_player_10', 'home_player_11', 'away_player_1', 
                         'away_player_2', 'away_player_3', 'away_player_4', 'away_player_5', 
                         'away_player_6', 'away_player_7', 'away_player_8', 'away_player_9', 
                         'away_player_10', 'away_player_11'], axis=1)

# removing information that we can't know before the game (many nulls here anyway)
final_df = final_df.drop(['goal', 'shoton', 'shotoff', 'foulcommit', 'card', 'cross', 'corner', 'possession'],
                         axis=1)

# Id doesen't give us anything
final_df = final_df.drop(['id'], axis=1)
# same as previous. Seems totally useless
final_df = final_df.drop(['match_api_id'], axis=1)


# merging country table with match table
final_df = final_df.merge(country_t, left_on='country_id', right_on='id').drop(['country_id', 'id'], axis=1)
final_df = final_df.rename(index=str, columns={"name": "country"})

# merging league table with match table
final_df = final_df.merge(league_t, left_on='league_id', right_on='id')
final_df = final_df.drop(['league_id', 'id', 'country_id'], axis=1)
final_df = final_df.rename(index=str, columns={"name": "league"})

# not sure what I should do with stage, date, season
final_df = merge_match_and_team(final_df, team_attributes_t)

# dropping some more unneeded columns
final_df.drop(["season", "stage", "date"], axis=1, inplace=True)
final_df.head()
```
